### 測試爬取搜尋結果第一頁

```python
URL = "https://udndata.com/ndapp/Searchdec?udndbid=udndata&page=1&SearchString=%AA%C0%BD%D7%2B%A4%E9%B4%C1%26gt%3B%3D20210101%2B%A4%E9%B4%C1%26lt%3B%3D20221231%2B%B3%F8%A7%4F%3D%C1%70%A6%58%B3%F8&sharepage=20&select=1&kind=2"

response = requests.get(URL)
response.encoding = "big5"

response.status_code 
# 200: Success
# 401: Unauthorized
# 404: Not Found
```

```python
soup = BeautifulSoup(response.content,"html.parser")
print(soup.prettify())
```

### 爬取所有"社論"新聞連結

```python
# 測試進度條
for i in tqdm(range(10)):
    time.sleep(1)
```

```python
link = []
date = []

for page in tqdm(range(23,41)): #搜尋結果共800篇，每頁呈現20筆結果
    
    URL = "https://udndata.com/ndapp/Searchdec?udndbid=udndata&page="+str(page)+"&SearchString=%AA%C0%BD%D7%2B%A4%E9%B4%C1%26gt%3B%3D20210101%2B%A4%E9%B4%C1%26lt%3B%3D20221231%2B%B3%F8%A7%4F%3D%C1%70%A6%58%B3%F8&sharepage=20&select=1&kind=2"
    response = requests.get(URL)
    response.encoding = "big5"
    soup = BeautifulSoup(response.content,"html.parser")
    
    # 取得20篇新聞搜尋結果
    news20 = soup.find("div", class_="list")

    # 篩選標籤為"社論"的新聞日期與其index(20篇文章中的第幾篇)
    add_index = []
    add_date = []
    index = 0
    for cate in news20.find_all("span", class_="source"):
        if "社論" in cate.text:
            add_index.append(index)
            add_date.append(cate.text.split("．")[0])
        index += 1
        
    # 擷取"社論"新聞連結
    add_link = []
    for index in add_index:
        add_link.append("https://udndata.com"+news20.find_all("a")[index].get("href"))
    
    # 合併
    link += add_link
    date += add_date
    
    time.sleep(60)
```

```python
empty = list(np.repeat(float("nan"),len(link)))
# index 0 to 22
data=pd.DataFrame(list(zip(link,date,empty,empty)),columns=["link","date","title","content"])
data
```

```python
# index 23 to 40
data2=pd.DataFrame(list(zip(link,date,empty,empty)),columns=["link","date","title","content"])
```

```python
data.to_csv("datalinks.csv", index=False) #encoding="utf-8"
data2.to_csv("datalinks2.csv", index=False) #encoding="utf-8"
```

```python
data_all = pd.concat([data,data2],ignore_index=True)
data_all
```

```python
data_all.to_csv("datalinks_all.csv", index=False)
```

### 測試爬取三篇"社論"新聞標題與內文

```python
driver = webdriver.Chrome("chromedriver")

for thelink in tqdm(data["link"][0:3]):

    URL = thelink
    driver.get(URL)
    driver.implicitly_wait(5)
    if driver.find_element(By.XPATH, "//div[@class='title']").text =="會員登入":
        driver.find_element(By.XPATH, "//a[@class='button-a-color']").click() # Find "定址登入" 鈕按下
    soup = BeautifulSoup(driver.page_source,"html.parser")
    add_title = soup.find("div", class_="story-title").find("h1").text
    add_content = soup.find("article").text.replace("\n", "")
    
    if add_title and add_content:
        link_index = data.index[data["link"]==thelink].tolist()[0]
        data.iloc[link_index,2] = add_title
        data.iloc[link_index,3] = add_content
    
    else:
        print("Error in id "+str(link_index)+": "+thelink)

    time.sleep(30)

driver.quit()

data
```
